```python
from datasets import load_dataset


df = load_dataset("lsb/million-english-numbers")
```

```python
X = df['train']['text'][:]
X = list(X)
y = list(range(0, 1000000))
```

```python
X[99999:100000]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
Y = []
for word in y:
  Y.append(word)
x = []
for word in X:
  x.append(word.lower())
```

```python
tokenizer = Tokenizer()
```

```python
y = []
for ele in Y:
    digits = [int(x) for x in str(ele)]
    y.append(digits)
```

```python

y = [[1] + num + [2] for num in y]
```

```python

output_tokenizer = tokenizer.fit_on_texts(x)
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python


import ast

y = [str(ele) for ele in y]
y = [ast.literal_eval(seq) for seq in y]
```

```python
x = tokenizer.texts_to_sequences(x)
```

```python
y = pad_sequences(y,padding='post',maxlen=9)
x = pad_sequences(x,padding='post',maxlen=9)
```

```python
y[:5]
```

```python
from tensorflow.keras.layers import Dense,Embedding,LSTM,Input
```

```python
output_vacab = len(tokenizer.word_index) + 1
output_vacab
```

```python
encoder_input = Input(shape=(None,))
encoder_emd = Embedding(10,50)(encoder_input)
encoder_lstm = LSTM(50,return_state=True)
_,h,c = encoder_lstm(encoder_emd)

decoder_input = Input(shape=(None,))
decoder_emd = Embedding(31,50)(decoder_input)
decoder_lstm = LSTM(50,return_sequences=True,return_state=True)
decoder_output,_,_ = decoder_lstm(decoder_emd,initial_state=[h,c])

decoder_dense = Dense(31,activation='softmax')
output = decoder_dense(decoder_output)
```

```python
from tensorflow.keras.models import Model
```

```python
model = Model([encoder_input,decoder_input],output)
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```
